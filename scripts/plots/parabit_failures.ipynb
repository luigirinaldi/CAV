```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import math
import json


benches = ["ROVER", "Hydra", "Alive", "Cadence"]

RESULTS_DIR = Path("../results")

parabit = {}

for bench in benches:
    parabit[bench] = pd.read_csv(f"{RESULTS_DIR}/parabit/{bench}/results.csv")
    parabit[bench]["time_taken"] = (
        parabit[bench]["time_taken"] * 1000
    )  # convert seconds to milliseconds
    parabit[bench]["file"] = parabit[bench]["file"].str.replace(
        ".bwlang", "", regex=False
    )  # remove file extension

for name, bench in parabit.items():
    fails = bench[bench['status'] != 'SUCCESS']
    print(f'# {name}', len(fails))
    errs = set(fails['last_err_line'])
    for err in errs:
        if '124' in err:
            string = 'timeout'
        elif '6' in err:
            string = 'memory'
        elif '101' in err:
            string = 'crash'
        elif '1' in err:
            string = 'saturation'
            # print(fails[fails['last_err_line'] == err]['file'])
        else:
            string = err
        print(string, len(fails[fails['last_err_line'] == err]))
        # print(fails[fails['last_err_line'] == err][['time_taken', 'max_memory_mb']])
```
